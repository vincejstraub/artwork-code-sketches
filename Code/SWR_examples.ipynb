# Demo: How to emulate _Walking nowhere in particular_ 

```python
from matplotlib import colors as mcolors
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import random 
import glob 
```

This notebook showcases examples of random walk visualizations generated using the script _SWRdrawer_, used to create the work _Walking nowhere in particular_ (2020). At each step in the random walk, the location is programmed to jump to a neighbouring site according to a predefined probability distribution and step count (see the function `fill_array` for details). 

## Single panel example
To generate a figure, you simply have to instantiate SRWdrawer. Here I set some of the arguments, such as figure size, rather than relying on the default values, so that the figure is not too large. 

```python
# Load simple random walk class from file
from SRWdrawer import SRWdrawer

# Instantiate SWRdrawer and show generated image
srw=SRWdrawer(steps=1000, fig_size= (4, 3), cols=1, rows=1)
srw.show_fig()
```

## Example with 8 panels
Here I generate 15 figures, setting the step count to increase exponentially, and set `face_color = 'black'` so that all figures have the same background. 

```python
# Generate list of exponentially increasing step values
step_vals = [2**exp for exp in range(0, 15)]

# Generate list of figure numbers
fig_nums = [1 + n for n in range(0, 15)]

# Combine step values and figure numbers in dictionary
my_dict = dict(zip(fig_nums, step_vals))

# Generate images
for fig_num, step in my_dict.items():
    srw=SRWdrawer(steps=step, face_color='black')
    # Save images to file
    srw.save_fig(file_num=fig_num)
    plt.close()
print('\n All images saved to file. Showing last image...')
srw.save_fig()
```
